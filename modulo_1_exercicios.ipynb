## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```python
# (A)
qtd_total_vendas_a = 3
tkt_a = 320.52
valor_total_vendas_a = tkt_a * qtd_total_vendas_a
print(valor_total_vendas_a)
```

```python
# (B)
valor_total_vendas_b = 834.47
tkt_b = 119.21
qtd_total_vendas_b = valor_total_vendas_b // tkt_b
print(qtd_total_vendas_b)
```

```python
# (C)
valor_total_vendas_c = 15378.12
qtd_total_vendas_c = 5
tkt_c = valor_total_vendas_c / qtd_total_vendas_c
print(tkt_c)
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```python
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

print(cancao.find("roda gigante"))
print(cancao.index("roda moinho"))
print(cancao.replace("roda pião","rodapé"))
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```python
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

print(noticia.find("2,75%"))
print(noticia.find("6"))

selic = noticia[12:17]
ano = noticia[62]
print(selic)
print(ano)
```

## 4\. Booleanos

Utilize a tabela da verdade para responder: qual o valor da variável x?

```python
a = False
b = True

x = not a & b

print(x)

print("A = False e B = True")
print("Tabela Verdade E (&):")

print(True & True)
print(True & False)
print(False & True)
print(False & False)

print(not True & True)
print(not True & False)
print(not False & True)
print(not False & False)
```
